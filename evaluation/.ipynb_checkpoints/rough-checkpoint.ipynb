```python

import plotly.graph_objects as go
import numpy as np
import math

# format of the columns in file name_of_image,error(l1),target,predicted

err_paths= ['nwpu_unbinned.txt','nwpu_rr.txt','nwpu_rs.txt']
bins = [0.0, 0.5, 2.5, 5.5, 10.5, 18.5, 31.5, 53.5, 90.5, 152.5, 256.5, 429.5, 717.5, 1197.5, 1997.5, 3330.5, 5552.5, 9255.5, 15426.0]


path = err_paths[0]
f = open(path, "r")
ub =[]
for x in f:
    x=x.replace('\n','').split(',')
    x =[float(i) for i in x]
    ub.append(x)

path = err_paths[1]
f = open(path, "r")
rr =[]
for x in f:
    x=x.replace('\n','').split(',')
    x =[float(i) for i in x]
    rr.append(x)


path = err_paths[2]
f = open(path, "r")
rs =[]
for x in f:
    x=x.replace('\n','').split(',')
    x =[float(i) for i in x]
    rs.append(x)


rr = np.asarray(rr)
rs = np.asarray(rs)
ub = np.asarray(ub)
  

def make_bins(bins):
    fin =[]
    for i in range(len(bins)-1):
        fin.append("("+str(bins[i])+" , "+str(bins[i+1])+")")
    return fin

def mean_bins(dm,bins):
    new =[]
    mean_std = []
    for i in range(len(bins)-1):
            if i ==0: # first bin
                mask = np.logical_and(dm[:,2]>=bins[i],dm[:,2] <=bins[i+1])
                mae = np.mean(np.abs(dm[mask]))
                std = np.std(np.abs(dm[mask]))
                mean_std.append([round(mae,3),round(std,3)])
            else: 
                mask = np.logical_and(dm[:,2]>bins[i],dm[:,2] <=bins[i+1])
                mae = np.mean(np.abs(dm[mask]))
                std = np.std(np.abs(dm[mask]))
                mean_std.append([round(mae,3),round(std,3)])
            new.append(sum(mask))
    mean_std = np.asarray(mean_std)
    return mean_std

def counts_in_bins(dm,bins):
    new =[]
    for i in range(len(bins)-1):
        if i ==0:
            mask = np.logical_and(dm[:,2]>=bins[i],dm[:,2] <=bins[i+1])
        elif i< len(bins)-1:
            mask = np.logical_and(dm[:,2]>bins[i],dm[:,2] <=bins[i+1])
            
        new.append(sum(mask))
    return new



ranges = make_bins(bins)
x_ax = np.linspace(0,len(bins),num = len(bins)+1)
unbinned = mean_bins(ub,bins)
counts= counts_in_bins(ub,bins)


def nonzero_bins(ranges,counts):
  modif_counts =counts.copy()
  modif_ranges = ranges.copy()
  to_keep = []
  m_r = []
  m_c = []
  for i in range(len(counts)-1):
    if i==len(counts)-1 and counts[i] == 0:
      break
    curr_range =  modif_ranges[i].replace("(","").replace(")","").replace(" ","").split(",")
    next_range = modif_ranges[i+1].replace("(","").replace(")","").replace(" ","").split(",")
    
    if counts[i] == 0:
      
      f = modif_ranges[i+1]
      a = next_range[0]
      b = curr_range[0]
     
      modif_ranges[i+1] = f.replace(a,b) 
    else:
        to_keep.append(i)
  if counts[-1] != 0:
    to_keep.append(len(counts)-1)
  for x in to_keep:
    m_r.append(modif_ranges[x])
    m_c.append(modif_counts[x])

  return to_keep,m_c,m_r

to_keep,m_c,m_r = nonzero_bins(ranges,counts)

def human_format(num, round_to=2):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num = round(num / 1000.0, round_to)
    return '{:.{}f}{}'.format(round(num, round_to), round_to, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])


def arr_to_sc(m_r):
    mm_r =[]
    #   seen =[]
    all_s =[]
    for i in range(len(m_r)):
        row = m_r[i]
        split = row.replace("(","").replace(")","").split(",")
        split[0] = float(split[0])
        split[1] = float(split[1])

        s = split[0]
        r = split[1]

        s = float(s)+0.1
        s = round(s,0)
        s = int(s)

        r =  float(r)-0.1
        r = round(r,0)
        r = int(r)

        s = human_format(s, round_to=1)
        r = human_format(r, round_to=1)
        if "K" not in s:
            s = int(float(s))
        if "K" not in r:
            r = int(float(r))


        t = "("+str(s)+" , "+str(r)+")"
        # print(t)
        mm_r.append(t)
        all_s.append(s)

  
    return mm_r

def pool_mean_std(txt,bins,to_keep,m_r,m_c):
  # to_keep,m_r,m_c = 
  means = mean_bins(txt,bins)[to_keep][:,0]
  stds =  mean_bins(txt,bins)[to_keep][:,1]

  pooled_mean = 0
  pooled_std = 0
  sum_ni = 0
  for i in range(len(m_c)):
    pooled_mean += (means[i]*m_c[i])
    pooled_std += (stds[i]**2 * m_c[i])
    sum_ni += m_c[i]

  pooled_mean = pooled_mean / sum_ni

  pooled_std = math.sqrt(pooled_std/sum_ni)
  return pooled_mean, pooled_std
```

```python
# individual sample plot

fig = go.Figure(layout=go.Layout(title=go.layout.Title(text="Per-sample performance of DM Count on NWPU dataset"),
                                 xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Number of people in a test image (gt) ")),
                                 yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Absolute error per test image"))))

x = np.arange(0,len(mean_bins(ub,bins)[to_keep][:,0])) 
    
fig.add_trace(go.Scatter(
        x=ub[:,2],mode ="markers",
        y=np.abs(ub[:,1]),name ='No-binning',textposition = "bottom center"))

fig.add_trace(go.Scatter(
        x=rs[:,2],mode ="markers",
        y=np.abs(rs[:,1]),name ='Bin Loss'))

fig.add_trace(go.Scatter(
        x=rr[:,2],mode ="markers",
        y=np.abs(rr[:,1]),name ='Bin Loss (RR) '))



# fig.update_xaxes(tickangle = 40,
#     ticktext=arr_to_sc(m_r),
#     tickvals=x,
# #     ticks="inside",tickson="boundaries"
# )
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)',xaxis=dict(showgrid=True),
                  yaxis=dict(showgrid=True,gridcolor='rgb(211,211,211)'),
    legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="right",
    x=0.89
), font=dict(size=15,color="#000000"),margin=dict(l=20 , r=20, t=40, b=20))

fig.update_layout(
    autosize=False,
    width=1000,
    height=600
)

fig.show()
```

```python
# individual sample plot

fig = go.Figure(layout=go.Layout(title=go.layout.Title(text="Per-bin performance of DM Count on NWPU dataset"),
                                 xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="bin range")),
                                 yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Bin wise MAE"))))

# x = np.arange(0,len(mean_bins(ub,bins)[to_keep][:,0])) 
    
fig.add_trace(go.Scatter(
        x=x,mode ="markers",
        y=mean_bins(ub,bins)[to_keep][:,0],name ='No-binning',textposition = "bottom center",
        error_y=dict(
            type='data', # value of error bar given in data coordinates
            array=mean_bins(ub,bins)[to_keep][:,1],
            visible=True)
    ))
fig.add_trace(go.Scatter(
        x=x+0.25,mode ="markers",
        y=mean_bins(rs,bins)[to_keep][:,0],name ='Bin Loss',
        error_y=dict(
            type='data', # value of error bar given in data coordinates
            array=mean_bins(rs,bins)[to_keep][:,1],
            visible=True)
    ))

fig.add_trace(go.Scatter(
        x=x+0.5,mode ="markers",
        y=mean_bins(rr,bins)[to_keep][:,0],name ='Bin Loss (RR)',
        error_y=dict(
            type='data', # value of error bar given in data coordinates
            array=mean_bins(rr,bins)[to_keep][:,1],
            visible=True)
    ))



fig.update_xaxes(tickangle = 40,
    ticktext=arr_to_sc(m_r),
    tickvals=x,
#     ticks="inside",tickson="boundaries"
)
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)',xaxis=dict(showgrid=True),
                  yaxis=dict(showgrid=True,gridcolor='rgb(211,211,211)'),
    legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="right",
    x=0.89
), font=dict(size=15,color="#000000"),margin=dict(l=20 , r=20, t=40, b=20))

fig.update_layout(
    autosize=False,
    width=1000,
    height=600
)

fig.show()
```
